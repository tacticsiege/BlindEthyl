```python
import sys
sys.path.append('C:\\git\\BlindEthyl\\RetinopathyProject\\')

from BlindEthyl.ImagePreprocessor import ImagePreprocessor
processor = ImagePreprocessor()
```

```python
import cv2
import numpy as np
from os import listdir
from os.path import isfile, join

path = 'samples\\'
files = [f for f in listdir(path) if isfile(join(path, f))]
imgs = []
for filename in files:
    imgs.append(cv2.imread(join(path, filename)))

output = []
for image in imgs:
    output.append(ImagePreprocessor.transform(image))
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (12.0, 10.0)
```

```python
out_image = np.concatenate(output, axis=1)
plt.imshow(out_image)
```
